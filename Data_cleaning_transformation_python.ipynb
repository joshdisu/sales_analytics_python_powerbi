```python
#check for duplicates 
duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print("Duplicate rows:", duplicate_rows)
else:
    print ("No duplicate rows")
```

```python
#add column names 
column_names = ['SalesOrderNumber', 'SalesOrderLine', 'OrderDate', 'CustomerName', 'EmailAddress', 'Item', 'Quantity', 'UnitPrice', 'TaxAmount']
```

```python
#read the csv file into a pandas df with the specified column name
df = pd.read_csv(r"C:\Users\joshu\OneDrive\Documents\Python Scripts\2021.csv", header = None, names = column_names)
```

```python
df.head()
```

```python
#Divide "Order Date" column by "year" and "month" columns
df[['Year', 'Month', 'Day']] = df['OrderDate'].str.split('-', expand = True)
```

```python
df.head()
```

```python
#Insert the new "year" and "month" columns after the "order date" column

#Get the index of the "Order Date" column
order_date_index = df.columns.get_loc('OrderDate')

#Insert "Year" column after "Order Date"
df.insert(order_date_index + 1,'Year', df.pop('Year'))
df.insert(order_date_index + 2,'Month', df.pop('Month'))
```

```python
df.head()
```

```python
#Drop the "day" column
df.drop('Day', axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Split the "CustomerName" column into "FirstName" and "LastName" columns
df[['FirstName', 'LastName']] = df['CustomerName'].str.split(' ', n = 1, expand = True)
df.head()
```

```python
#Insert the LastName and FirstName columns after the "CustomerName" column
Customername_index = df.columns.get_loc('CustomerName')

#Insert "Year" column after "Order Date"
df.insert(Customername_index + 0,'FirstName', df.pop('FirstName'))
df.insert(Customername_index + 0,'LastName', df.pop('LastName'))

df.head()
```

```python
df.drop('CustomerName', axis = 1, inplace = True)

df.head()
```

```python
df.to_csv('python_transformed_dataset.csv', index= False)
```
